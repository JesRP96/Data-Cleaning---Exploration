```python
import pandas as pd
```

```python
csv_file = "data/DataBreaches(2004-2021).csv"

df = pd.read_csv(csv_file)

df_modified = df.copy()# Creating a copy
```

```python
# How many companies where hacked?
hacked = df_modified[df_modified["Method"]=="hacked"]
hacked["Entity"].nunique() # Nunique shows the amount of uniques per column
```

```python
# How many organization types are there?
uniqueOrgs = len(df_modified["Organization type"].unique())
print("There are " + str(uniqueOrgs) + " unique organization types")
```

```python
# How may types of methods are there and what are those?
uniqueMethods = pd.DataFrame(df_modified["Method"].unique())
#df_modified[df_modified["Method"].unique()]
print("There are " + str(len(uniqueMethods)) + " unique methods")
uniqueMethods
```

```python
# Which organization type had the most breaches?
sumByOrg = df_modified.groupby("Organization type")["Records"].sum().reset_index()
# When grouping you need to specify by which column you will group, later which columns you want and then specify which aggregation statement you will perform on the query

sumByOrg = sumByOrg.sort_values(by="Records", ascending = False)
sumByOrg.to_csv("sumByOrg.csv",index=0)
```

```python
# Which organization type was breached the least?
leastBreachedType = sumByOrg.sort_values(by="Records")
```

```python
# Which web company was breached the most?
webMostBreached = df_modified[df_modified["Organization type"]=="web"].sort_values(by="Records",ascending=False)
```

```python
# How many records were breached for the most breached organization type (web) by method
webMostRecordsbyMethod = df_modified[df_modified["Organization type"]=="web"].sort_values(by="Records",ascending=False).groupby("Method")["Records"].sum().reset_index().sort_values("Records",ascending=False)
```

```python
# Which advertising companies were breached?
df_modified[df_modified["Organization type"]=="advertising"]
```

```python
# Which company had the most records breached?
mostOrgBreached = df_modified.sort_values(by="Records",ascending=False)
mostOrgBreached.head(1)
```

```python
# Which company had the least records breached?
leastBreachedOrg = df_modified.sort_values("Records").head(1)
leastBreachedOrg
```

```python
# Which year was the worst for breaches?
recordsbyYear = df_modified.groupby("Year")["Records"].sum().reset_index().sort_values(by="Records",ascending=0).head(5)
recordsbyYear
```

```python
# Which year had the least breaches? (removing ambigous years)
df_mod1 = df_modified[df_modified["Year"]!="2014 and 2015"]
df_mod1 = df_mod1[df_mod1["Year"]!="2018-2019"]
df_mod1.groupby("Year")["Records"].sum().reset_index().sort_values("Records")
```

```python
# How many companies were breached by method?
df_modified["Method"].value_counts().reset_index()
```

```python
# Replacing the "poor security" inconsistency
df_modified["Method"] = df_modified["Method"].str.replace("Poor security","poor security")
```

```python
# Which company(s) had accidental exposure as a breach?
df_modified[df_modified["Method"]=="accidentally exposed"]
```

```python
# Did some companies had multiple breaches?
multipleBreaches = df_modified["Entity"].value_counts().reset_index()
multipleBreaches
```
